In this part, one of the common imputation method is implemented. Which is to replace missing values with the *mean* of the corresponding **feature** (column) across the available data points. This can be represented mathematically as:
Let $X$ be the original dataset with missing values.
Let $X_{imputed}$ be the imputed dataset after replacing missing values.
For each feature $j$, the imputation process can be represented as:
Compute the mean of the available values in feature $j$ as $\mu_j$.
Replace missing values in feature $j$ with $\mu_j$.
